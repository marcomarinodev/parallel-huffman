```python
import matplotlib.pyplot as plt

# Data - Speedup Comparison
speedup_data = {
    '10MB': {
        'threads': [2, 4, 8, 16, 32, 64],
        'native_threads': [1.617, 3.017, 4.082, 5.550, 9.373, 8.527],
        'fastflow': [1.946, 3.340, 3.917, 3.629, 4.827, 4.649]
    },
    '50MB': {
        'threads': [2, 4, 8, 16, 32, 64],
        'native_threads': [1.462, 2.179, 3.696, 5.012, 5.935, 6.405],
        'fastflow': [1.638, 2.431, 3.686, 4.263, 3.293, 3.612]
    },
    '100MB': {
        'threads': [2, 4, 8, 16, 32, 64],
        'native_threads': [1.419, 2.201, 3.314, 4.454, 6.670, 7.064],
        'fastflow': [1.567, 2.543, 3.848, 3.558, 4.135, 3.513]
    }
}

# Data - Scalability
scalability_data = {
    '10MB': {
        'threads': [2, 4, 8, 16, 32, 64],
        'native_threads': [1.730, 3.596, 4.868, 5.962, 11.152, 10.177],
        'fastflow': [1.524, 2.618, 3.067, 2.844, 3.774, 3.642]
    },
    '50MB': {
        'threads': [2, 4, 8, 16, 32, 64],
        'native_threads': [1.829, 3.147, 4.951, 6.697, 8.557, 8.523],
        'fastflow': [1.532, 2.678, 3.446, 3.975, 3.679, 3.606]
    },
    '100MB': {
        'threads': [2, 4, 8, 16, 32, 64],
        'native_threads': [1.842, 3.751, 4.502, 7.195, 9.417, 10.254],
        'fastflow': [1.552, 2.427, 3.678, 3.400, 3.947, 3.357]
    }
}

# Set Greyscale Color Map
greyscale = plt.get_cmap('Greys')

# Plot - Speedup Comparison
for size, data in speedup_data.items():
    plt.figure(figsize=(8, 5))
    plt.plot(data['threads'], data['native_threads'], marker='o', color=greyscale(0.7), label='Native Threads')
    plt.plot(data['threads'], data['fastflow'], marker='o', color=greyscale(0.4), label='FastFlow')
    plt.xlabel('Number of Threads')
    plt.ylabel('Speedup')
    plt.title(f'Speedup Comparison - {size}')
    plt.legend()
    plt.show()

# Plot - Scalability
for size, data in scalability_data.items():
    plt.figure(figsize=(8, 5))
    plt.plot(data['threads'], data['native_threads'], marker='o', color=greyscale(0.7), label='Native Threads')
    plt.plot(data['threads'], data['fastflow'], marker='o', color=greyscale(0.4), label='FastFlow')
    plt.xlabel('Number of Threads')
    plt.ylabel('Scalability')
    plt.title(f'Scalability - {size}')
    plt.legend()
    plt.show()
```
